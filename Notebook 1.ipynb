## Predicting house prices using multiple linear regression

In this project we are gonna see how Machine Learning algoritms help us predict house prices. Linear  Regresion is a model of predicting a new future data by using the existing correlation between the old data. Here, Machine Learning help us identify this relationship between future data and output, so we can predict feature values.

```
import pandas as pd
from sklearn import linear_model
```

```
df = pd.read_csv('propiedades.csv', sep = ';')
df
```

### The following is our feature set

```
df[['area', 'roomcount', 'buildingage']]
```

### The following is the output(result) data

```
df['price']
```

```
reg = linear_model.LinearRegression()
reg.fit(df[['area','roomcount', 'buildingage']], df['price'])
```

### Since our model is ready, we can make predictions now

Lets predict a house with 230 square meters, 4 rooms and 10 years old building

```
reg.predict([[230,4,10]])
```

Now lets predict a house with 230 square meters, 6 rooms and 0 year old building

```
reg.predict([[230, 6,0]])
```

Now lets predict a house with 355 square meters, 3 rooms and 20 years old building

```
reg.predict([[355,3,20]])
```

You con make as many predictions as you want

```
reg.predict([[230,4,10], [230,6,0], [355,3,20], [275,5,17]])
```

### Now we´ll see the coefficients of our multilinear regression formula

```
reg.coef_
```

```
reg.intercept_
```

```
a = reg.intercept_
b1 = reg.coef_[0]
b2 = reg.coef_[1]
b3 = reg.coef_[2]

x1 = 230
x2 = 4
x3 = 10
y = a + b1*x1 + b2*x2 + b3*x3

y
```
